```python
# Element-wise addition

a = [i for i in range(10000000)]
b = [i for i in range(20000000)]

c = []

import time

start = time.time()

for i in range(len(a)):
    c.append(a[i] + b[i])

print(time.time() - start)
```

# Broadcasting 

* Used in Vectorization 

The term broadcasting describes how NumPy treats arrays with different shapes during arithmetic operations.

The smaller array is "broadcast" across the larger array so that they have compatible shapes.

```python
# same shape 

a = np.arange(6).reshape(2,3)
b = np.arange(6,12).reshape(2,3)

print(a)
print(b)

print(a+b)
```

```python
# diff shape 
a = np.arange(6).reshape(2,3)
b = np.arange(3).reshape(1,3)

print(a)
print(b)

print(a+b)
```

# Broadcasting  Rules 
# 1.Make the two arrays have the same number of dimensions 

* If the numbers of dimensions of the two arrays are different,add new dimensions with size 1 to the head of the array with the smaller dimension.
* Ex:(3,2) = 2D,(3)=1D--->Convert into(1,3)
(3,3,3)=3D,(3)=1D---> Convert into(1,1,3)

# 2.Make each dimensions of the two arrays the same size 

* If the sizes of each dimensions of the two arrays do not match , dimensions with size 1 are stretched to the size of the other array .

* Ex:(3,3)=2D,(3)--->CONVERTED (1,3)than stretch to (3,3)

* If there is a dimension whose size is not 1 in either of the two arrays , it cannot be broadcast , and an error is raised 

```python
# More Examples 

a = np.arange(12).reshape(4,3)
b = np.arange(3)

print(a) # 2D
```

```python
print(b) # 1D
```

```python
print(a+b) # Arithmetic Operation
```

# Numpy Meshgrid Creates Coordinates for a Grid System 

np.meshgrid ([1,2,3][5,6,7])

* These arrays , xv and yv , each seperately give the x and y coordinates on a 2D grid . You can do normal numpy operations on these arrays:

```python
xv**2 + yv**2
```

This can be done on a larger scale to plot surface plots of 2D functions 

Generate functions f(x,y)=e-(x^2+y^2) for -2_<x_<2 and -1 _< y _<1

```python
x = np.linspace(-2,2,100)
y = np.linspace(-1,1,100)
xv,yv = np.meshgrid(x,y)
f = np.exp(-xv**2-yv**2)
```

Note:pcolormesh is typically the preferable function for 2D plotting , as opposed to imshow or pcolor , which take longer.)

```python
plt.figure(figsize=(6,3))
plt.pcolormesh(xv,yv,f,shading='auto')
plt.colorbar()
plt.grid()
plt.show()
```

f(x,y)=1&x^2+y^2\0&x^2+y^2

```python
import numpy as np 
import matplotlib .pyplot as plt 

def f(x,y):
   return np.where ((x**2 + y**2 < 1),1.0,0.0)

x = np.linspace(-5,5,500)
y = np.linspace(-5,5,500)
xv , yv = np.meshgrid(x,y)
rectangular_mask = f(xv, yv)

plt.pcolormesh(xv,yv,rectangular_mask , shading='auto')
plt.colorbar()
plt.grid()
plt.show()
```
