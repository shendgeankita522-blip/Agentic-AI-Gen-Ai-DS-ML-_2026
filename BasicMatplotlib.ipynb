# Figure Class
* Example
Here is an example that uses the pyplot.figure() method to create a figure

```python
import matplotlib.pyplot as plt 
import numpy as np 

# Creating the figure instance 
fig = plt.figure(figsize=[7,30], facecolor ='lightgreen',layout='constrained')

# Adding a title to the figure 
fig.suptitle('Figure')

# Adding a title to the figure
ax = fig.add_subplot()

# Setting a title for the subpllot
ax.set_title('Axes', loc = 'left', fontstyle='oblique', fontsize='medium')

# Showing the plot 
plt.show()
```

# Matplotlib - Setting Ticks and Tick Labels 

```python
import matplotlib.pyplot as plt 
import numpy as np 

x = np.arange(0,10,1)
y = np.sin(x)

plt.plot(x,y)
plt.xticks([0,2,4,6,8],['Zero','Two','Four','Six','Eight'])
plt.yticks([-1,0,1],['Min','Zero','Max'])
plt.title('Sine Wave')
plt.show()
                     
```

# 1.set_xticklabels() and set_yticklabels()

```python
import matplotlib.pyplot as plt 

x = [0,1,2,3,4,5,6]
y = [0,-1,0,1,0,-1,0]

fig, ax = plt.subplots()
ax.plot(x,y)

ax.set_xticks([0,2,4,6])
ax.set_yticks([-1,0,1])

ax.set_xticklabels(['Zero','Two','Four','Six'])
ax.set_yticklabels(['MIn','Zero','Max'])

plt.show()
```

# Multiple Plots 
* matplotlib.pyplot.subplot2grid(shspe,loc,rowspan=1,colspan=1,fig=None,**kwargs)

```python
# Importing libraries 
import matplotlib.pyplot as plt 
import numpy as np 
import math

# Placing the plots in the plane 
plot1 = plt.subplot2grid((3,3),(0,0),colspan=2)
plot2 = plt.subplot2grid((3,3),(0,2),rowspan=3,colspan=2)
plot3 = plt.subplot2grid((3,3),(1,0),rowspan=2)

# Using Numpy to create an array X
x = np.arange(1,10)

#Plot for square root
plot2.plot(x,x**0.5)
plot2.set_title('Square Root')

# Plot for exponent
plot1.plot(x,np.exp(x))
plot1.set_title('Exponent')

# Plot for square 
plot3.plot(x,x*x)
plot3.set_title('Square')

#packing all the plots and displaying them 
plt.tight_layout()
plt.show()
```

# Line Graph 

```python
# Importing the rquired libraries 
import matplotlib.pyplot as plt 
import numpy as np 

# define data values
x = np.array([1,2,3,4]) # x-axis points 
y = x*2 # y-axis points 

plt.plot(x,y) # plot the chart
plt.show() # display 
```

# Bar Chart

```python
# define array 
data = [5.,25.,50.,20.]
plt.bar(range(len(data)),data,color='c')
plt.show()
```

# Histograms

```python
# draw random samples from random distribution 
x = np.random.normal(180,20,350)

#plot histograms
plt.hist(x)
plt.show()
```

# Scatter Plot

```python
# create the x and y axis coodinates 
x = np.array([5,7,2,4,8,9,11,2,34,89,90,2,3,56,87])
y = np.array([99,110,120,10,18,17,77,88,98,65,60,39,34,47,56])
plt.scatter(x, y)
plt.legend()
plt.show()
```

# Pie chart 

```python
# define the figure size 
plt.figure(figsize=(8,8))
x =[35,40,55,15]

# Labels of the pie chart 
labels = ['1','2,','3','4']
plt.pie(x,labels=labels)
plt.show()
```

# Box Plots

```python
# Create the random values by using numpy 
values = np.random.normal(200,30,400)

# creating the plot by boxplot() function which is available in matplotlib 
plt.boxplot(values,patch_artist=True,vert=True)
plt.show()
```
