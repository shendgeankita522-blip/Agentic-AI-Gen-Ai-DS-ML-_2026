```python
# python is not vectorize programming language
# Building dashboards (dashboard - combination of chats)

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')

f, axes = plt.subplots(2, 2, figsize=(15, 15))

# Top-left: Budget vs Audience
k1 = sns.kdeplot(
    data=movies,
    x='BudgetMillions',
    y='AudienceRating',
    ax=axes[0, 0],
    fill=True,
    cmap='Greens'
)

# Top-right: Budget vs Critic
k2 = sns.kdeplot(
    data=movies,
    x='BudgetMillions',
    y='CriticRating',
    ax=axes[0, 1],
    fill=True,
    cmap='Blues'
)

k1.set(xlim=(-20, 160))
k2.set(xlim=(-20, 160))

# Bottom-left: Violin plot (Drama only)
z = sns.violinplot(
    data=movies[movies.Genre == 'Drama'],
    x='Year',
    y='CriticRating',
    ax=axes[1, 0]
)

# Bottom-right: Critic vs Audience density
k4 = sns.kdeplot(
    data=movies,
    x='CriticRating',
    y='AudienceRating',
    fill=True,
    cmap='Reds',
    ax=axes[1, 1]
)

# Optional overlay (contour lines only)
k4b = sns.kdeplot(
    data=movies,
    x='CriticRating',
    y='AudienceRating',
    color='black',
    ax=axes[1, 1]
)

# Titles (important for dashboard clarity)
axes[0, 0].set_title('Budget vs Audience')
axes[0, 1].set_title('Budget vs Critic')
axes[1, 0].set_title('Drama: Critic Rating by Year')
axes[1, 1].set_title('Critic vs Audience Density')

plt.tight_layout()
plt.show()
```
