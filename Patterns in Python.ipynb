# 1)Right Angle Triangle Pattern

```python
for i in range(1,6):
    print(i)
```

# 2) Inverted Right Angle Triangle Pattern

```python
for i in range(5,0,-1):
    print(i)
```

# 3) Pyramid Pattern

```python
for i in range (1,6):
    print(" " * (5 - i) + " * " * (2 * i - 1))
```

# 4) Inverted Pyramid Pattern

```python
for i in range (5, 0, -1):
    print( "  " * (5 - i) + " * " * (2 * i - 1))
```

# 5) Diamond Pattern

```python
for i in range (1,6):
    print( " " * (5 - i) + " * " * (2 * i -1))
    for i in range (4,0,-1):
        print( " " *(5 - i) + " * " * (2 * i - 1))
```

# 8) Right Angle traingle (number pattern)

```python
for i in range (1,6):
    print( '  '.join(str(x) for x in range (1, i + 1)))
```

# 9) Inverted Right Angle Traingle (Number Pattern)

# 10) Floyd's Traingle

```python
num = 1 
for i in range(1,6):
    for j in range(1 ,i + 1):
        print(num,end=' ')
        num += 1
    print()
```

# 11) Hallow Right Angle Traingle

```python
for i in range(1,6):
    for j in range(1, i + 1):
        if j  == 1 or j == i or i == 5:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

# 12) Hallow Pyramid Pattern

```python
n = 5

for i in range(1, n + 1):
    for j in range(1, 2 * n):
        if i == n or i + j == n + 1 or j - i == n - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

# 13) Hallow Diamond Pattern

```python
n = 5 

# Upper half 
for i in range(1, n + 1):
    # spaces 
    for j in range(n - i):
        print(" ",end=" ")

    # numbers
    for j in range (2 * i - 1):
        if j==0 or j == 2 * i - 2:
            print(i,end=" ")
        else:
            print(" ",end=" ")
    print()

# lower half
for i in range(n - 1, 0 , -1):
    #spaces
    for j in range(n - i):
        print(" " , end= " ")


    # numbers
    for j in range(2 * i - 1):
        if j == 0 or j == 2 * i - 2:
            print(i,end=" ")
        else:
            print(" " , end=" ")
    print()
            
```

# 14) Butterfly patterns

```python
n = 5

# Upper half
for i in range(1, n + 1):
    # left numbers
    for j in range(1, i + 1):
        print(j, end=" ")

    # spaces
    for j in range(2 * (n - i)):
        print(" ", end=" ")

    # right numbers
    for j in range(1, i + 1):
        print(j, end=" ")

    print()

# Lower half
for i in range(n, 0, -1):
    # left numbers
    for j in range(1, i + 1):
        print(j, end=" ")

    # spaces
    for j in range(2 * (n - i)):
        print(" ", end=" ")

    # right numbers
    for j in range(1, i + 1):
        print(j, end=" ")

    print()
```

```python
n = 5

# Upper half
for i in range(1, n + 1):
    for j in range(i):
        print("*", end=" ")

    for j in range(2 * (n - i)):
        print(" ", end=" ")

    for j in range(i):
        print("*", end=" ")

    print()

# Lower half
for i in range(n - 1, 0, -1):
    for j in range(i):
        print("*", end=" ")

    for j in range(2 * (n - i)):
        print(" ", end=" ")

    for j in range(i):
        print("*", end=" ")

    print()
```

# 15) Hallow Diamond (Number Pattern)

```python
n = 5

# Upper half
for i in range(1, n + 1):
    for j in range(n - i):
        print(" ", end=" ")

    for j in range(1, 2):
        print(i, end=" ")

    for j in range(2 * i - 3):
        print(" ", end=" ")

    if i != 1:
        print(i, end=" ")

    print()

# Lower half
for i in range(n - 1, 0, -1):
    for j in range(n - i):
        print(" ", end=" ")

    print(i, end=" ")

    for j in range(2 * i - 3):
        print(" ", end=" ")

    if i != 1:
        print(i, end=" ")

    print()
```

# 16)Hallow Number Pyramid

```python
n = 5

for i in range(1, n + 1):
    # spaces
    for j in range(n - i):
        print(" ", end=" ")
    
    # pattern (numbers or stars)
    for j in range(1, 2 * i):
        if j == 1 or j == 2 * i - 1 or i == n:
            print(i, end=" ")
        else:
            print(" ", end=" ")
    
    print()
```

# 17)Full Star Pyramid 

```python
n = 5

for i in range(1, n + 1):
    for j in range(n - i):
        print(" ", end=" ")
    for j in range(2 * i - 1):
        print("*", end=" ")
    print()
```

# 18) Inverted Full Star Pyramid

```python
n = 5

for i in range(n, 0, -1):
    for j in range(n - i):
        print(" ", end=" ")
    for j in range(2 * i - 1):
        print("*", end=" ")
    print()
```

# 19) Left Aligned Pyramid(Stars)

```python
n = 5

for i in range(1, n + 1):
    for j in range(i):
        print("*", end=" ")
    print()
```

# LEFT Aligned Number Pattern 

```python
n = 5

for i in range(1, n + 1):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

# 20) Right Aligned Pyramid (Numbers)

```python
n = 5

for i in range(1, n + 1):
    for j in range(n - i):
        print(" ", end=" ")
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```
