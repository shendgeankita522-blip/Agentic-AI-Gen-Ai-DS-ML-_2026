# Data Cleaning or Data Cleansing

```python
emp['Name']
```

```python
emp['Name'] = emp['Name'].str.replace(r'\W','',regex=True) # non word character
```

```python
emp['Name']
```

```python
emp
```

```python
emp['Domain']
```

```python
emp['Domain'] = emp['Domain'].str.replace(r'\W','')
```

```python
emp['Domain']
```

```python
emp
```

```python
emp['Age'] = emp['Age'].str.replace(r'\W','')
emp['Age']
```

```python
emp["Age"]=emp["Age"].str.replace(r'[^0-9]','',regex=True)
emp["Age"]
```

```python
emp['Age'] = emp['Age'].str.extract('(\d+)')
emp['Age']
```

```python
emp
```

```python
emp['Location'] = emp['Location'].str.replace(r'\W','')
emp['Location']
```

```python
emp['Salary']
```

```python
emp['Salary'] = emp['Salary'].astype(str).str.replace(r'[^0-9]', '', regex=True)
```

```python
emp['Salary'] = emp['Salary'].str.replace(r'\W','')
emp['Salary']
```

```python
emp
```

```python
emp['Exp']
```

```python
emp['Exp'] = emp['Exp'].str.extract('(\d+)')
emp['Exp']
```

* Till now we have raw data we use regex to clean the data and removed all noise characterd from the dataset

# EDA techniques lets apply 

* Missing values treatment for numerical data 

```python
clean_data = emp.copy()
clean_data
```

```python
clean_data.dtypes
```

```python
clean_data['Age']
```

```python
import numpy as np 
```

```python
clean_data['Age'] = clean_data['Age'].fillna(np.mean(pd.to_numeric(clean_data['Age'])))
clean_data['Age']
```

```python
clean_data['Exp']
```

```python
clean_data['Exp'] = clean_data['Exp'].fillna(np.mean(pd.to_numeric(clean_data['Exp'])))
clean_data['Exp']
```

```python
clean_data
```

```python
clean_data['Location'].isnull().sum()
```

```python
clean_data['Location']
```

```python
clean_data['Location'] = clean_data['Location'].fillna(clean_data['Location'].mode()[0])
clean_data['Location']
```

```python
clean_data
```

```python
clean_data.info()
```

```python
clean_data['Age'] = clean_data['Age'].astype(int)
```

```python
clean_data.info()
```

```python
clean_data['Salary'] = clean_data['Salary'].astype(int)
clean_data['Exp'] = clean_data['Exp'].astype(int)
```

```python
clean_data.info()
```

```python
clean_data['Name'] = clean_data['Name'].astype('category')
clean_data['Domain'] = clean_data['Domain'].astype('category')
clean_data['Location'] = clean_data['Location'].astype('category')
```

```python
clean_data.info()
```

```python
clean_data
```

```python
clean_data.to_csv('clean_data.csv')
```
